```python
import numpy as np
import matplotlib.pyplot as plt
```

Transition rates
\begin{equation}
\mathcal{\omega_{21}} ({t}) = \mu cos^2(\omega\ t)
\end{equation}
\begin{equation}
\mathcal{\omega_{12}} ({t}) = \mu sin^2(\omega\ t)
\end{equation}

```python
mu = 4
frequency = 4

rate1_list = []
rate2_list = []

x_axis = np.linspace(0,5, 1000)

for t in x_axis:
    rate1 = mu*(np.cos(frequency*t))**2
    rate2 = mu*(np.sin(frequency*t))**2
    rate1_list.append(rate1)
    rate2_list.append(rate2)
    

plt.figure(figsize =(15,4))
plt.plot(x_axis, rate1_list, label = '$ω_{21}$')
plt.plot(x_axis, rate2_list, label = '$ω_{12}$')

plt.ylabel('$ω$')
plt.xlabel('time')
plt.legend(loc='upper right')
plt.savefig('hw2_num5a', dpi = 300)
plt.show()
```
